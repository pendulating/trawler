```python
def chat_gpt(prompt, temperature=0):
    response = client.chat.completions.create(
        model="gpt-5",
        messages=prompt
    )
    return response.choices[0].message.content.strip()
```

```python
def replace_key(d, old_key, new_key):
  """
  Replace `old_key` with `new_key` in dictionary `d`.
  The associated value is retained.
  """
  if old_key in d:
      d[new_key] = d.pop(old_key)
  return d
```

```python
def format_prompt(MESSAGES, domain,purpose,aiCapability,aiUser,aiSubject,space):
    S = "test {}"
    messages = deepcopy(MESSAGES)
    messages[1]['content'] = messages[1]['content'].format(domain,purpose,aiCapability,aiUser,aiSubject,space)
    return messages
```

```python
def parse_single_use(useElements,MESSAGES):

    start_time = time.time()

    useI = str(useElements['UseID'])
    print (f" Parsing use {useI}")

    # Variables for message placeholder
    domain = useElements['Domain']
    purpose = useElements['Purpose']
    aiCapability = useElements['Capability']
    aiUser = useElements['AI_Deployer']
    aiSubject = useElements['AI_Subject']
    space = useElements['Space']
    # location = useElements['Location']

    # Extracting "Use i" details
    # use_i_details = [domain, purpose, aiCapability, aiUser, aiSubject, space]
    # print(use_i_details)

    # adapt the prompt for useI
    messages = format_prompt(MESSAGES, domain,purpose,aiCapability,aiUser,aiSubject,space)
    response = chat_gpt(messages)
    print(response)
    end_time = time.time()
    print (f"This use processing took {end_time-start_time} seconds.")


    return response
```

# Batch Processing


https://cookbook.openai.com/examples/batch_processing

```python

# Creating an array of json tasks

tasks = []

for useElements in USES:
    
    useI = str(useElements['UseID'])
    print (f" Parsing use {useI}")

    # Variables for message placeholder
    domain = useElements['Domain']
    purpose = useElements['Purpose']
    aiCapability = useElements['Capability']
    aiUser = useElements['AI_Deployer']
    aiSubject = useElements['AI_Subject']
    space = useElements['Space']

    # adapt the prompt for useI
    messages7s = format_prompt(MESSAGES, domain,purpose,aiCapability,aiUser,aiSubject,space)
    
    task = {
        "custom_id": f"task-{int(useI)}",
        "method": "POST",
        "url": "/v1/chat/completions",
        "body": {
            # This is what you would have in your Chat Completions API call
            "model": "gpt-5",
            "response_format": { 
                "type": "json_object"
            },
            "messages": messages7s,
        }
    }
    
    tasks.append(task)
```

## Save Batch File

```python
# Creating the file

file_name = "../../processing/batch_impact_of_AI_uses.jsonl"

with open(file_name, 'w') as file:
    for obj in tasks:
        file.write(json.dumps(obj) + '\n')
```

```python
results_benefits = []
results_risks = []

for i, tmp_el in results.iterrows():
  # print (tmp_el)
  tmp = tmp_el["Assessment of impact on human rights"]
  # print (tmp)
  tmp_benefits_el = {}
  tmp_risks_el = {}

  for use in USES:
    if str(use["Representative IncidentID"]) == str(tmp_el["Incident ID"]):
      domain = use['Domain']
      purpose = use['Purpose']
      aiCapability = use['Capability']
      aiUser = use['AI User']
      aiSubject = use['AI Subject']
      space = use['Space']
      break

  
  tmp_benefits_el["Domain"] = domain
  tmp_benefits_el["Purpose"] = purpose
  tmp_benefits_el["Capability"] = aiCapability
  tmp_benefits_el["AI User"] = aiUser
  tmp_benefits_el["AI Subject"] = aiSubject
  tmp_benefits_el["Space"] = space 

  tmp_risks_el["Domain"] = domain
  tmp_risks_el["Purpose"] = purpose
  tmp_risks_el["Capability"] = aiCapability
  tmp_risks_el["AI User"] = aiUser
  tmp_risks_el["AI Subject"] = aiSubject
  tmp_risks_el["Space"] = space  

  tmp_benefits_el["Representative IncidentID"] = tmp_el["Incident ID"]
  tmp_risks_el["Representative IncidentID"] = tmp_el["Incident ID"]

  tmp_benefits_el["Assessment of impact on human rights"] = \
    [article for i, article in enumerate(tmp) if f"Article {i+1}" in article and article[f"Article {i+1}"] in ["Positive", "Mixed"] ]
  results_benefits.append(tmp_benefits_el)

  tmp_risks_el["Assessment of impact on human rights"] = \
    [article for i, article in enumerate(tmp) if f"Article {i+1}" in article and article[f"Article {i+1}"] in ["Negative", "Mixed"] ]
  results_risks.append(tmp_risks_el)
```
